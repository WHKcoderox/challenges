```python
'''
Guess the number game.
The computer selects a random number between 1 and 100.
The user keeps guessing which number the computer has chosen
until they get it right.
The computer responds 'bingo!' or 'too high' or 'too low' after
each guess.
After the user has guessed the number correctly the computer output
how many attempts it has made.
'''
import random
import datetime

random.seed(datetime.datetime.now())
num = random.randint(1,100)
count = 0
while True:
    user_input = int(input("Guess the number (1-100): "))
    if user_input > 100 or user_input < 1:
        print("Invalid guess")
        continue
    else:
        count += 1
    if user_input > num:
        print("Too high")
    elif user_input < num:
        print("Too low")
    else:
        print("You win! {0} tries taken.".format(count))
        break
```

```python
'''
Reverse the above game. You think of a number. Get the
computer to guess a number and you respond with too high(H),
too low(L) or bingo!(B). Make sure the computer has a game
plan - don't just let it guess at random!
'''
# binary search
while True:
    user_input = int(input("Set a number between 1-100, and the computer will try to guess!"))
    if user_input > 100 or user_input < 0:
        print("Invalid number")
        continue
    else:
        print("You picked", user_input)
        break
high = 100
low = 1
count = 0
while True:
    mid = (high + low)//2
    print("Computer guesses", mid)
    user_input = input("Is this too high(H), too low(L) or bingo(B)?\n")
    if user_input.upper() == "B":
        count += 1
        print("Yay! The computer wins after {0} tries!".format(count))
        break
    elif user_input.upper() == "H":
        high = mid - 1
        count += 1
    elif user_input.upper() == "L":
        low = mid + 1
        count += 1
    else:
        print("what?")
```

```python
'''
Write a function that will convert a mark into a grade.
The function will return 'A', 'B', 'C', 'D', 'E', 'S' or 'U'.
The function will require a parameter mark.
The formula is >=70% â> 'A', >=60% â>'B', >=55% â> 'C',
>=50% â> 'D', >=45% â>'E', >=40% â> 'S', <40% â> 'U'.
Assume the maximum mark is 100.
Having written the function we want to use it three times.
Write a program with the function that allows the user to enter
two module marks and displays the grade. It then computes the
average and displays the overall grade. E.g.

Enter Module 1 mark: 68
Enter Module 2 mark: 57
Result
Module 1: B
Module 2: C
Overall: B 
'''
m1 = float(input("Enter Module 1 mark: "))
m2 = float(input("Enter Module 2 mark: "))
final = (m1 + m2) / 2
grades = [m1,m2,final]
for i in range(3):
    if grades[i] >= 70.0:
        grades[i] = 'A'
    elif grades[i] >= 60.0:
        grades[i] = 'B'
    elif grades[i] >= 55.0:
        grades[i] = 'C'
    elif grades[i] >= 50.0:
        grades[i] = 'D'
    elif grades[i] >= 45.0:
        grades[i] = 'E'
    elif grades[i] >= 40.0:
        grades[i] = 'S'
    else:
        grades[i] = 'U'
print("Result\nModule 1: {0}\nModule 2: {1}\nOverall: {2}".format(grades[0], grades[1], grades[2]))  
```

```python
'''
Write a program to perform a basic encryption and decryption on text.
This algorithm works by moving letters along by an offset. 
If the offset is 2 then b â> d, h â> j, etc.
Write two functions - encrypt and decrypt. Both will return a string.
The user selects whether to encrypt or decrypt.
The user enters sentence to encrypt and the offset.
The program responds with the encrypted or decrypted version. 
'''
alphabet = [char for char in "abcdefghijklmnopqrstuvwxyz"]
def encrypt(string, offset):
    string = string.lower()
    new_string = []
    for char in string:
        symbol = True
        for i in range(26):
            if char == alphabet[i]:
                new_string.append(alphabet[(i+offset)%26])
                symbol = False
        if symbol:
            new_string.append(char)
    return ''.join(new_string)

def decrypt(string, offset):
    string = string.lower()
    new_string = []
    for char in string:
        symbol = True
        for i in range(26):
            if char == alphabet[i]:
                new_string.append(alphabet[(i-offset)%26]) # can always use a conditional to make this one function instead
                symbol = False
        if symbol:
            new_string.append(char)
    return ''.join(new_string)
while True:
    user_input = input("Enter 'encrypt' to encrypt and 'decrypt' to decrypt.\n")
    if user_input.lower() == 'encrypt':
        user_input = input("Enter text you want to " + user_input + " followed by offset(no spacing) [i.e. blablabla,2]:\n")
        user_input = user_input.split(',')
        print(encrypt(user_input[0], int(user_input[1])))
        break
    elif user_input.lower() == 'decrypt':
        user_input = input("Enter text you want to " + user_input + " followed by offset(no spacing) [i.e. blablabla,2]:\n")
        user_input = user_input.split(',')
        print(decrypt(user_input[0], int(user_input[1])))
        break
    else:
        print("Invalid selection.")
```

```python
'''
The above algorithm is actually easy to crack. A much better algorithm 
would be one that has a different offset for every letter. 
We can do this using the random number generator - because it generates 
the same sequence of random numbers from a 'seed' position. As long as
sender and receiver agree where to 'seed' (this will be the encryption
key) they can both work out the same offsets. 

Algorithm for encrypt function

function Encrypt(text, key)
  Randomize(key)
  For each letter in text
    Get its ascii code
    Add random offset int(rnd()*10) to ascii code
    Turn this new ascii code back to a character
    Append character to ciphertext string
  Endfor
  return ciphertext
end
'''
import random
def encrypt(text, key):
    random.seed(key)
    new_text = []
    for letter in text:
        new_text.append(chr(ord(letter) + random.randint(1,10)))
    return ''.join(new_text)
def decrypt(text, key):
    random.seed(key)
    new_text = []
    for letter in text:
        new_text.append(chr(ord(letter) - random.randint(1,10)))
    return ''.join(new_text)
```

```python
'''
For the above task, use a single function with an extra parameter to indicate
whether the text is being encrypted or decrypted rather than having two 
different functions.
'''
def crypt(text, key, encrypt):
    random.seed(key)
    new_text = []
    for letter in text:
        if encrypt:
            r = random.randint(1,10)
        else:
            r = -random.randint(1,10)
        new_text.append(chr(ord(letter) + r))
    return ''.join(new_text)
```
