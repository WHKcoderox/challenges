```python
''' 
Write a program for a game where the computer generates a
random starting number between 20 and 30.
The player and the computer can remove 1,2 or 3 from the number
in turns. Something like this:

Starting number: 25
How many do you want to remove? 3
22 left
Computer removes 2
20 left
...

The player who has to remove the last value to bring the number
down to 0 is the loser.

1 left
Computer removes 1
You win!

Easy: Get the computer to choose a number between 1â3 at random. 
'''
import random
import datetime
random.seed(datetime.datetime.now())
starting_num = random.randint(21,29)
print("Starting number:", starting_num)
while starting_num > 1:
    random.seed(datetime.datetime.now())
    user_num = int(input("How many do you wannt to remove? (1,2 or 3) "))
    if starting_num < user_num or user_num <= 0 or user_num > 3:
        print("Unfortunately you cannot do that")
        continue
    starting_num -= user_num
    if starting_num == 0:
        print("You lose :(")
        break
    print(starting_num, "left")
    if starting_num < 3:
        cp_num = random.randint(1,starting_num)
    else:
        cp_num = random.randint(1,3)
    starting_num -= cp_num
    print("Computer removes", cp_num)
    if starting_num == 0:
        print("You win!")
    else:
        print(starting_num, "left")
```

```python
''' 
Write a program for a game where the computer generates a
random starting number between 20 and 30.
The player and the computer can remove 1,2 or 3 from the number
in turns. Something like this:

Starting number: 25
How many do you want to remove? 3
22 left
Computer removes 2
20 left
...

The player who has to remove the last value to bring the number
down to 0 is the loser.

1 left
Computer removes 1
You win!

Hard: Get the computer to employ a strategy to try and win. 
'''
random.seed(datetime.datetime.now())
starting_num = random.randint(21,29)
print("Starting number:", starting_num)
while starting_num > 0:
    random.seed(datetime.datetime.now())
    # user turn
    user_num = int(input("How many do you wannt to remove? (1,2 or 3) "))
    if starting_num < user_num or user_num <= 0 or user_num > 3:
        print("Unfortunately you cannot do that")
        continue
    starting_num -= user_num
    if starting_num == 0:
        print("You lose :(")
        break
    print(starting_num, "left")
    # cp turn
    cp_num = (starting_num - 1)%4 # 1 + 4n are losing numbers
    if cp_num == 0:
        # losing, maximise chances by randomising
        cp_num = random.randint(1,3)
    starting_num -= cp_num
    print("Computer removes", cp_num)
    if starting_num == 0:
        print("You win!")
    else:
        print(starting_num, "left")
```

```python
3####
'''
Write a program for a Higher / Lower guessing game.
The computer randomly generates a sequence of up to 10 numbers
between 1 and 13. The player after seeing each number in turn
has to decide whether the next number is higher or lower.
If you get 10 guesses right you win the game.

Starting number : 12
Higher(H) or lower(L)? L
Next number 8
Higher(H) or lower(L)? L
Next number 11
You lose

Hints
- Use a condition controlled loop (do until, while etc) to control the
game.
- You do not need to remember all 10 numbers, but  just the current and
the next number.
- Remember to keep a count of the number of guesses. 
'''
import random
count = 0
user_guess = "H"
prev = 0
while True:
    if count == 10:
        print("You win!")
        break
    starting_num = 0
    while starting_num == prev or starting_num == 0:
        starting_num = random.randint(1,13)
    if count == 0:
        print("Starting", end="")
    else:
        print("Next", end='')
    print(" number: {0}".format(starting_num))
    if user_guess == "H" and starting_num < prev:
        print("You lose")
        break
    elif user_guess == "L" and starting_num > prev:
        print("You lose")
        break
    else:
        prev = starting_num
    while True:
        user_guess = input("Higher(H) or lower(L)? ")
        if user_guess != "H" and user_guess != "L":
            print("Invalid guess. Type 'H' for higher and 'L' for lower.")
            continue
        else:
            break
    count += 1
```

```python
'''
Extend the above Higher / Lower number guessing game by
- Giving the player two lives
- Allowing only H or L to be entered 
'''
import random
count = 0
user_guess = "H"
prev = 0
lives = 2
while True:
    if count == 10:
        print("You win!")
        break
    starting_num = 0
    while starting_num == prev or starting_num == 0:
        starting_num = random.randint(1,13)
    if count == 0:
        print("Starting", end="")
    else:
        print("Next", end='')
    print(" number: {0}".format(starting_num))
    if user_guess == "H" and starting_num < prev:
        lives -= 1
        print("You lose a life. Lives left:", lives)
    elif user_guess == "L" and starting_num > prev:
        lives -= 1
        print("You lose a life. Lives left:", lives)
    else:
        prev = starting_num
    if lives == 0:
        print("You lose.")
        break
    while True:
        user_guess = input("Higher(H) or lower(L)? ")
        if user_guess != "H" and user_guess != "L":
            print("Invalid guess. Type 'H' for higher and 'L' for lower.")
            continue
        else:
            break
    count += 1
```

```python
'''
Write a program to count the number of words in a sentence.
The user enters a sentence.
The program outputs the number of words in the sentence.
Hint: Look for spaces and full stops in the string. 
'''
sentence = input("Type in any sentence. It must end in either a full stop(.), exclamation mark(!) or question mark(?).\n Words like \"shouldn't\" or \"short-form\" count as two words.\n")
sentence = sentence.split()
count = 0
for word in sentence:
    count += 1
    for i in range(len(word)):
        if i == "-" or i == "'":
            count += 1
print("Word count:", count)
```
