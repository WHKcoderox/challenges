```python
# Write a program to work out how many days you have lived for.
import datetime
bday = input("Enter birth date in DDMMYYYY format(i.e. 13th May 2010 as 13052010):\n")
b_d = int(bday[0:2])
b_m = int(bday[2:4])
b_y = int(bday[4:])
b_date = datetime.datetime(b_y,b_m,b_d)
delta = datetime.datetime.now() - b_date
print("Days lived:", delta.days)
```

```python
# Write a program that will give the answer to logic gate questions e.g.
# Enter logic gate : OR
# Enter first input : 1
# Enter second input :0
# Result = 1
# It should work for the logic gates OR, XOR, NAND and NOR gates.
while True:
    in_0 = input("Enter logic gate (OR, XOR, NAND, NOR). Enter 'stop' to stop:\n")
    l_gates = "OR,XOR,NAND,NOR".split(',')
    if in_0.lower() == "stop":
        break
    if in_0 not in l_gates:
        print("Invalid logic gate")
        continue
    in_1 = bool(int(input("Enter first input : ")))
    in_2 = bool(int(input("Enter second input : ")))
    if in_0 == l_gates[0]:
        if in_1 or in_2:
            print("Result = 1")
            continue
        else:
            print("Result = 0")
    elif in_0 == l_gates[1]:
        if (in_1 and not in_2) or (in_2 and not in_1):
            print("Result = 1")
            continue
        else:
            print("Result = 0")
    elif in_0 == l_gates[2]:
        if in_1 and in_2:
            print("Result = 0")
            continue
        else:
            print("Result = 1")
    elif in_0 == l_gates[3]:
        if in_1 or in_2:
            print("Result = 0")
            continue
        else:
            print("Result = 1")
```

```python
# Write a program that will display all the factors of an input number that
# are bigger than 1.
# (e.g. the factors of the number 12 are 6,4,3 and 2 because they divide
# into 12 exactly).
# Hint: To find out whether a number X is a factor of Y use:
# If Y mod X == 0 (there is nothing remaining when Y is divided by X)
in_num = input("Give a number > 1:\n")
try:
    in_num = int(in_num)
    if(in_num <= 1):
        print("Nope.")
    else:
        print("The factors are ", end="")
        for i in range(2, in_num):
            if in_num%i==0:
                print(i, end=", ")
        print(end="\b\b.\n")
except:
    print("Not a number")
```
